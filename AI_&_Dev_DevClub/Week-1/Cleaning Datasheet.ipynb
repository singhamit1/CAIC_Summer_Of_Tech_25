```python
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path="/content/behaviour_simulation_train.xlsx - Sheet1.csv"
ds=pd.read_csv(path)
ds.head(5)
```

```python
ds['content']
```

```python
ds['date'] = pd.to_datetime(ds['date'])
```

```python
ds.head(5)
```

```python
ds.dropna(subset=['content', 'username', 'inferred company', 'likes'], inplace=True)
```

```python
ds['media'].fillna('no_media', inplace=True)
ds['has_media'] = ds['media'].apply(lambda x: x != 'no_media')
ds['content'] = ds['content'].astype(str).str.strip().str.lower()
```

```python
sns.histplot(ds['likes'], kde=True)
sns.boxplot(x=ds['likes'])
```

```python
ds['hour'] = ds['date'].dt.hour
ds['day_of_week'] = ds['date'].dt.day_name()
```

```python
ds['word_count'] = ds['content'].apply(lambda x: len(x.split()))
ds['char_count'] = ds['content'].apply(len)
```
